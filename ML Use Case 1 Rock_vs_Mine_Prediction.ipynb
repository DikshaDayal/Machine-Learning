```python
#Importing the Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
#loading the dataset to a pandas Dataframe
sonar_ds = pd.read_csv('/content/sonar data.csv')
```

```python
sonar_ds.head()
```

```python
#Dimensions of the dataset
sonar_ds.shape
```

```python
#Describe the statistical Measures
sonar_ds.describe()
```

```python
sonar_ds['R'].value_counts()
```


M --> Mine

R --> Rock

```python
sonar_ds.groupby('R').mean()
```

```python
#Seperating Data and Labels
Features = sonar_ds.drop(columns='R' , axis='1')
Outcomes = sonar_ds['R']
```

```python
print(Features)
```

```python
print(Outcomes)
```

Trainning and Test data

```python
Features_train , Features_test , Outcomes_train  , Outcomes_test = train_test_split(Features , Outcomes , test_size= 0.1 , stratify = Outcomes, random_state= 2)
```

```python
print(Features.shape , Features_train.shape , Features_test.shape)
```

```python
print(Features_train)
print(Outcomes_train)
```

Model Training --> Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model.fit(Features_train, Outcomes_train)
```


Model **Evaluation**

```python
#accuracy on training data
Features_train_prediction = model.predict(Features_train)
training_data_accuracy = accuracy_score(Features_train_prediction, Outcomes_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
Features_test_prediction = model.predict(Features_test)
test_data_accuracy = accuracy_score(Features_test_prediction, Outcomes_test)
```

```python
print('Accuracy on testing data : ',test_data_accuracy)
```
