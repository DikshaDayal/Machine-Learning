**Imbalanced Dataset**

A dataset with an unequal data distribution.

```python
#importing the dependecies
import numpy as np
import pandas as pd
```

```python
# Loading the dataset int pandas dataframe
Credit_Card = pd.read_csv('/content/credit_data.csv')
```

```python
#first five row of the Dataset
Credit_Card.head()
```

```python
#last five row of the Dataset
Credit_Card.tail()
```

```python
Credit_Card.isnull().sum()
```

```python
#distribution of the two classes
Credit_Card['Class'].value_counts()
```

This is Highly Imbalanced Dataset

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
#Seperate the legit and Fradulent Transactions
Legit_Transactions = Credit_Card[Credit_Card.Class == 0]
Fraud_Transactions = Credit_Card[Credit_Card.Class == 1]
```

```python
print(Legit_Transactions)
print(Fraud_Transactions)
```

```python
Legit_Transactions.shape
```

```python
Fraud_Transactions.shape
```

**UnderSampling**

Build a sample dataset containing similar distribution of Legit & Fraudulent Transactionds

Number of Fraudulent Transactions --> 81

```python
Legit_sample = Legit_Transactions.sample(n = 81)
```

```python
Legit_sample.shape
```

Concatinate the Data frame

```python
new_dataset = pd.concat([Legit_sample , Fraud_Transactions] , axis = 0)
```
