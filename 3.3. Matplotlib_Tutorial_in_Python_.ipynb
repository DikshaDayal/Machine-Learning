Matplotlib -----> (it is very useful for making plots and Graphs)
* useful for making Plot.

```python
#importing Matplot library
import matplotlib.pyplot as plt
```

```python
# import numpy library to get data for plots
import numpy as np
```

```python
x = np.linspace(0,10,100)

y = np.sin(x)
z = np.cos(x)
print(x)
print(y)
print(z)
```

Plotting the Data

```python
#sine wave
plt.plot(x,y)  #(x-> x-axis , y-> y-axis)
plt.show()  # This will show the plot
```

```python
#cosine wave
plt.plot(x,z)  #(x-> x-axis , z-> y-axis)
plt.show()
```

```python
#Adding title , x-label and y-label
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.title('Sine Wave')
plt.show()


# cosine wave
plt.plot(x,z)
plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.title('Cosine Wave')
plt.show()
```

```python
#Parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.xlabel('input values')
plt.ylabel('Output')
plt.title('Parabola')
plt.show()
```

```python
#Discrete plot
plt.plot(x,y,'r+')
plt.xlabel('Input values')
plt.ylabel('Magnitude')
plt.title('Parabola')

plt.show()
```

```python
plt.plot(x,y,'g.')
plt.xlabel('Input values')
plt.ylabel('Magnitude')
plt.title('Parabola')

plt.show()
```

```python
plt.plot(x,y,'yx')
plt.xlabel('Input values')
plt.ylabel('Magnitude')
plt.title('Parabola')

plt.show()
```

```python
x = np.linspace(-5,5,50)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y , 'g')
plt.plot(x,z , 'r--')
```

Bar plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.5,0.5])  #(Last two values denoted the dimension sof the bar graph)
languages = ['English' , 'French' , 'Spanish' , 'Latin' , 'German']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.xlabel('Number of People')
plt.show()
```

Pie chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English' , 'French' , 'Spanish' , 'Latin' , 'German']
people = [100,50,150,40,70]
ax.pie(people, labels=languages, autopct='%1.1f%%')  #Autopct will tell you the floating point of it
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
